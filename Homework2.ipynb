```python
import gurobipy as gp
from gurobipy import GRB
import numpy as np
```

```python
#Q1 solution 1
m = gp.Model('Q1')
y = np.array([2**i for i in range(6)])
x = m.addMVar(shape=(3,6), vtype='B', name="x")
x_total = np.array([gp.quicksum([y[j] * x[i,j] for j in range(6)]) for i in range(3)])
b = m.addMVar(shape = (1,3),vtype=GRB.BINARY, name="b")
bigM = 100000
for i in range(3):
    m.addConstr(x_total[i] <= bigM * b[0,i])
set_cost = gp.quicksum(b[0][j] * np.array([500,400,300])[j] for j in range(3))
profit = gp.quicksum(x_total[i] * np.array([52,30,20])[i] for i in range(3)) - set_cost
m.setObjective(profit, GRB.MAXIMIZE) 
m.addConstr(gp.quicksum(x_total * np.array([2,4,5])) <= 100)
m.addConstr(gp.quicksum(x_total * np.array([1,1,1])) <= 30)
m.addConstr(gp.quicksum(x_total * np.array([10,5,2])) <= 204)

m.optimize()
print()
print('Profit = ' + str(int(m.ObjVal)))
q = [0,0,0]
for i in range(3):
    for j in range(6):
            q[i] += y[j] * m.getVars()[i*6+j].x
print('Gold = ' + str(int(q[0])))
print('Silver = ' + str(int(q[1])))
print('Bronze = ' + str(int(q[2])))
```

```python
#Q1 solution 3
m = gp.Model('Q1')
y = np.array([2**i for i in range(6)])
x = m.addMVar(shape=(3,6), vtype='B', name="x")
x_total = np.array([gp.quicksum([y[j] * x[i,j] for j in range(6)]) for i in range(3)])
b = m.addMVar(shape = (1,3),vtype=GRB.BINARY, name="b1")
epsilon = 0.5
bigM = 100000
for i in range(1):
    for j in range(3):
        m.addConstr(x_total[j] >= epsilon - bigM * (1-b[i,j]))
        m.addConstr(x_total[j] <= bigM * b[i,j])
set_cost = gp.quicksum(b[0][j] * np.array([500,400,300])[j] for j in range(3))
profit = gp.quicksum(x_total * np.array([52,30,20])) - set_cost
m.setObjective(profit, GRB.MAXIMIZE) 
m.addConstr(gp.quicksum(x_total * np.array([2,4,5])) <= 100)
m.addConstr(gp.quicksum(x_total * np.array([1,1,1])) <= 30)
m.addConstr(gp.quicksum(x_total * np.array([10,5,2])) <= 204)

m.optimize()
print()
print('Profit = ' + str(int(m.ObjVal)))
q = [0,0,0]
for i in range(3):
    for j in range(6):
            q[i] += y[j] * m.getVars()[i*6+j].x
print('Gold = ' + str(int(q[0])))
print('Silver = ' + str(int(q[1])))
print('Bronze = ' + str(int(q[2])))
```

```python
#Q4 solution 
m = gp.Model('Q2')

prob = np.array([[0.5,0.6,0.7,0.5],[0.6,0.7,0.8,0.7],[0.8,0.8,0.9,0.9]]).T
cost = np.array([[1,2,1,2],[2,4,3,3],[3,5,4,4]]).T
x = m.addMVar(shape = (4,3),vtype='B', name="x")
#p01 = m.addVar(vtype=GRB.CONTINUOUS, name="p01")
#p23 = m.addVar(vtype=GRB.CONTINUOUS, name="p23")
p = m.addVar(vtype=GRB.CONTINUOUS, lb = 0, ub = 1, name="p")
x_0_m = m.addMVar(shape = (3,3),vtype='B', name="x_0_m")
x_1_m = m.addMVar(shape = (3,3),vtype='B', name="x_1_m")
x_0_1 = m.addMVar(shape = (3,3),vtype='B', name="x_0_1")
x_2_m = m.addMVar(shape = (3,3),vtype='B', name="x_2_m")
x_3_m = m.addMVar(shape = (3,3),vtype='B', name="x_3_m")
x_2_3 = m.addMVar(shape = (3,3),vtype='B', name="x_2_3")

prob_matrix_0_1 = np.ones((3,3))
for i in range(3):
    for j in range(3):
        prob_matrix_0_1[i,j] = prob[0,i] * prob[1,j]
prob_matrix_2_3 = np.ones((3,3))
for i in range(3):
    for j in range(3):
        prob_matrix_2_3[i,j] = prob[2,i] * prob[3,j]
       
# x_0_m = np.zeros((3,3))
# x_1_m = np.zeros((3,3))
for i in range(3):
    for j in range(3):
        m.addConstr(x_0_m[i,j] == x[0,i])
        m.addConstr(x_1_m[i,j] == x[1,j])
for i in range(3):
    for j in range(3):
        m.addConstr(x_0_1[i,j] == x_0_m[i,j] * x_1_m[i,j])
#x_2_m = np.zeros((3,3))
#x_3_m = np.zeros((3,3))
for i in range(3):
    for j in range(3):
        m.addConstr(x_2_m[i,j] == x[2,i])
        m.addConstr(x_3_m[i,j] == x[3,j])
for i in range(3):
    for j in range(3):
        m.addConstr(x_2_3[i,j] == x_2_m[i,j] * x_3_m[i,j])
        
#x_0_1 = np.zeros((3,3), dtype = np.int8)
#for i in range(3):
#    for j in range(3):
#        m.addConstr(x_0_1[i,j] == gp.quicksum(x[0,i] * x[1,j] for j in range(3) for i in range(3)))
#x_2_3 = np.zeros((3,3))
#for i in range(3):
#    for j in range(3):
#        m.addConstr(x_2_3[i,j] == gp.quicksum(x[2,i] * x[3,j] for j in range(3) for i in range(3)))
        
p_0_1 = gp.quicksum(prob_matrix_0_1[i,j] * x_0_1[i,j] for j in range(3) for i in range(3))
p_2_3 = gp.quicksum(prob_matrix_2_3[i,j] * x_2_3[i,j] for j in range(3) for i in range(3))

#p_0_1 = np.ones((3,3))
#for i in range(3):
#    for j in range(3):
#        p_0_1[i,j] = prob_matrix_0_1[0,i] * x_0_1[1,j]
#p_2_3 = np.ones((3,3))
#for i in range(3):
#    for j in range(3):
#        p_2_3[i,j] = prob_matrix_2_3[0,i] * x_2_3[1,j]

        
        
# p01 = gp.quicksum(x[0,j] * prob_matrix_0_1[0,j] for j in range(3))  *

# p = [0,0,0,0]
# for i in range(4):
#    p[i] = gp.quicksum(prob[0,j] * x[0,j] for j in range(3))

totalCost = 0
for i in range(4):
    totalCost += gp.quicksum(cost[i,j] * x[i,j] for j in range(3))

for i in range(4):
    m.addConstr(gp.quicksum(x[i]) == 1)
m.addConstr(totalCost <= 10)

m.addConstr(p == p_0_1 * p_2_3)
#m.addConstr(p01 == p[0] * p[1])
#m.addConstr(p23 == p[2] * p[3])
#m.addConstr(pAll == p01* p23)
m.setObjective(p, GRB.MAXIMIZE) 

m.optimize()
if m.status == GRB.Status.INFEASIBLE:
    m.computeIIS()
    m.write("model1.ilp")
print()
print('Obj: %g' % m.ObjVal)
print()
for v in m.getVars():
    print('%s %g' % (v.VarName, v.X))
```
